```python
import matplotlib.pyplot as plt

```

```python
plt.plot([4,5,6],[10,8,3])
plt.show()
plt.xlabel('india')
plt.ylabel('united kingdom')
plt.title('developed countries')
plt.xlim(10,2)
plt.ylim(15,3)
```

```python
x = [4,5,6,8,9,10]
y = [2,10,8,9,13,15]
plt.scatter(x,y,color='red')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
plt.legend()
```
